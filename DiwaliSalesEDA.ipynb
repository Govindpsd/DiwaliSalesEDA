# Exploratory Data Analysis(EDA)

**Gender**

```python
#Generating a graph of Male Female count using countplot
#Now if we want count to be displayed as a label on top of each bar we need to use a for loop 
#to traverse throught the data and count it
mf = sns.countplot(x='Gender',data = df)
for count in mf.containers:#Containers are used to store count as a tuple and using for loop inside it
#can help to retrieve all the data.
        mf.bar_label(count)
```

```python
#Now since we have calculated the count of both the genders now we will check purchase power of both the genders
#AND choose which one is the highest among two.
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot( x = 'Gender', y = 'Amount', data = sales_gen)
df.columns
```

From the bar it can be concluded that the purchase power of Females are more then that of males.

```python
#Here we checked the maximum purchaase power by each group
age_sales=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=age_sales)
```

```python
#Now to check male and female of same age group using hue
#here hue means that the count will be based on gender as well as u can see in the plot there are two genders.
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for count in ax.containers:
    ax.bar_label(count)
```

From the above graph we can say that maximum buyers are from 26-35 years age where females are maximum.

**State**

```python
#to check maximum number of orders from each state
st_ord=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})#this command is used to change barplot size
sns.barplot(x='State',y='Orders',data=st_ord)

```

From here we can see the maximum order are from up,mh,karnatka respectively

```python
#Finding the maximum amount used by states
st_am=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head()
sns.barplot(x='State',y="Amount",data=st_am)
df.columns
```

Also the maximum amount used are by UP,MH,KA respectively

**Marital Status(Shaadi)**

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(7,5)})
for count in ax.containers:
    ax.bar_label(count)
```

```python
ms_am=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=ms_am,hue='Gender')
```

Here we can see that maximum buyers are married (moslty females)

**Occupation**

```python
df.columns
#Now we are going to check count of people on each occupation.
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation',data=df)
for count in ax.containers:
    ax.bar_label(count)
```

```python
#Now we are going to check amount spend by each sector
sec_amt=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=sec_amt)
```

From here we can see that people on IT sector has the maximum amount spent

**Product Category**

```python
df.columns
#to check number of product categories
ax=sns.countplot(x='Product_Category',data=df)
for count in ax.containers:
    ax.bar_label(count)
```

```python
#Now checking maximum amount on products
sns.set(rc={'figure.figsize':(30,5)})
pd_amt=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Product_Category',y='Amount',data=pd_amt)
```

```python
df.columns
#Now checking orders done from product id.
pd_ord=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Orders',data=pd_ord)
```
